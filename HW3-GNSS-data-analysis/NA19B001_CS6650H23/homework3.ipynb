````python
from pathlib import Path

import numpy as np
import pandas as pd
from tabulate import tabulate

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns

sns.set_theme()
cmap = sns.color_palette()


class GnssLogScenario:
    # GNSS log Fix message headers
    FIX_HEADERS = [
        "Fix",
        "Provider",
        "LatitudeDegrees",
        "LongitudeDegrees",
        "AltitudeMeters",
        "SpeedMps",
        "AccuracyMeters",
        "BearingDegrees",
        "UnixTimeMillis",
        "SpeedAccuracyMps",
        "BearingAccuracyDegrees",
        "elapsedRealtimeNanos",
        "VerticalAccuracyMeters",
    ]

    # GNSS log Status message headers
    STATUS_HEADERS = [
        "Status",
        "UnixTimeMillis",
        "SignalCount",
        "SignalIndex",
        "ConstellationType",
        "Svid",
        "CarrierFrequencyHz",
        "Cn0DbHz",
        "AzimuthDegrees",
        "ElevationDegrees",
        "UsedInFix",
        "HasAlmanacData",
        "HasEphemerisData",
        "BasebandCn0DbHz",
    ]

    def __init__(self, name, infile, img_path) -> None:
        self.name = name
        self.logfile_all = Path(infile)
        self.img_path = Path(img_path)

        assert (
            self.logfile_all.exists() and self.img_path.exists()
        ), "One or more paths does not exist"

        self.df_fix: pd.DataFrame = None
        self.df_status: pd.DataFrame = None

    @property
    def logfile_fix(self):
        """Filter raw log file for `Fix` messages (only GPS) and dump the fix messages to a new file. Return filtered filepath directly if present."""

        # Generete `Fix` filtered filename
        stem, sfx = self.logfile_all.stem, self.logfile_all.suffix
        fix_file = self.logfile_all.parent / (stem + "-fix-filt" + sfx)

        # Return if file already exists
        if fix_file.exists():
            return fix_file

        # Filter `Fix` messages and write to file
        with open(self.logfile_all, "r") as f_in, open(fix_file, "w") as f_out:
            for line in filter(lambda l: l.startswith("Fix,GPS"), f_in):
                f_out.write(line.rstrip() + "\n")

        return fix_file

    @property
    def logfile_status(self):
        """Filter raw log file for `Status` messages and dump the status messages to a new file. Return filtered filepath directly if present."""

        # Generate `Status` filtered filename
        stem, sfx = self.logfile_all.stem, self.logfile_all.suffix
        status_file = self.logfile_all.parent / (stem + "-status-filt" + sfx)

        # Return if file already exists
        if status_file.exists():
            return status_file

        # Filter `Status` messages and write to file
        with open(self.logfile_all, "r") as f_in, open(status_file, "w") as f_out:
            for line in filter(lambda l: l.startswith("Status,"), f_in):
                f_out.write(line.rstrip() + "\n")

        return status_file

    def cleanup(self):
        """Delete all intermediate files created."""
        self.logfile_fix.unlink()
        self.logfile_status.unlink()
````

### Task 2: Compute and plot location errors using haversine distance [15 points] <a id='task2'></a>
Calculate the mean location coordinates for all three locations. Assume those three locations as the `groundtruth`. Compute the error for each location fix sample, which is the distance between the location sample and the groundtruth location. You must use haversine distance ([Wiki](https://en.wikipedia.org/wiki/Haversine_formula)). Plot the CDF (cumulative distribution function) of these errors for the three locations. Comment on the variance of the three distributions.


````python
def haversine_distance(lat1, lon1, lat2, lon2):
    """Calculate the Haversine distance between two points on the Earth's surface,
        given their latitudes and longitudes in degrees.

    Parameters:
    -----------
    lat1 : float
        Latitude of the first point in degrees.
    lon1 : float
        Longitude of the first point in degrees.
    lat2 : float
        Latitude of the second point in degrees.
    lon2 : float
        Longitude of the second point in degrees.

    Returns:
    --------
    distance_km : float
        The great-circle distance between the two points in kilometers,
        calculated using the Haversine formula.
    """
    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(np.deg2rad, [lat1, lon1, lat2, lon2])

    d_lat = lat2 - lat1
    d_lon = lon2 - lon1

    # Intermediate calculation for great-circle distance
    a = np.sin(d_lat / 2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(d_lon / 2) ** 2

    # Calculate great-circle distance in kilometers
    radius_earth_km = 6371
    central_angle = 2 * np.arcsin(np.sqrt(a))
    distance_km = radius_earth_km * central_angle

    return distance_km


for scn in scenarios:
    # Parse `Fix,GPS,...` messages
    scn.df_fix = pd.read_csv(
        scn.logfile_fix, header=None, names=GnssLogScenario.FIX_HEADERS
    )

    # Parse `Status,...` messages
    scn.df_status = pd.read_csv(
        scn.logfile_status, header=None, names=GnssLogScenario.STATUS_HEADERS
    )

    # For GNSS logs with limited visibility (indoor scenario) timestamps are sometimes not present leading to NaN values. Handle this.
    scn.df_status.UnixTimeMillis.fillna(np.inf, inplace=True)

    # Save mean of locations as ground truth
    mean_lat = scn.df_fix["LatitudeDegrees"].mean()
    mean_lon = scn.df_fix["LongitudeDegrees"].mean()
    scn.df_fix = scn.df_fix.assign(
        GtLatitudeDegrees=mean_lat, GtLongitudeDegrees=mean_lon
    )

    # Compute great circle distance between ground truth and each fix locationI
    scn.df_fix = scn.df_fix.assign(
        HDistErrorKm=haversine_distance(
            scn.df_fix["LatitudeDegrees"],
            scn.df_fix["LongitudeDegrees"],
            scn.df_fix["GtLatitudeDegrees"],
            scn.df_fix["GtLongitudeDegrees"],
        )
    )


# CDF of error values for all 3 scenarios
plt.figure(figsize=(6, 6), constrained_layout=True)
for i, scn in enumerate(scenarios):
    err = scn.df_fix["HDistErrorKm"].copy()
    hist, bins = np.histogram(err)
    hist_norm = hist / hist.sum()
    cdf = np.cumsum(hist_norm)

    plt.plot(bins[1:], cdf, label=scn.name.title(), color=cmap[i])

    plt.text(0.007, 0.5 - i * 0.05, rf"$\sigma^2 = {np.var(cdf):0.3f}$", color=cmap[i])

plt.title("CDF of Error Values Across Scenarios", fontweight="bold")
plt.ylabel("Cumulative Density")
plt.xlabel("Error (km)")
plt.legend()
plt.show()
````

**Approach**

- For each scenario
  - Mean latitude and longitude was computed and used henceforth as the ground truth.
  - The error (great circle distance) from the ground truth was calculated for every fix message using the _haversine formula_.
  - The Cumulative Distribution Function (CDF) of the error values were computed and visualised.
  - Variance of the distributions were computed and analysed.

**Observations**

### Task 3: Analyze status entries before fix and their relation to error [45 points] <a id='task3'></a>
Check the `status` entries prior to receiving a `fix`.
1. How many satellites were used for the fix (check the UsedInFix entry for the satellites) for the three locations? (state the median # of satellites)
2. How does the distance error for a fix correlate to the number of satellites used for that fix? Show scatter plots for the three locations. Summarize your observations.
3. How does the distance error for a fix correlate to the average SNR (Cn0DbHz entry) of the satellites that were used in the fix? Show scatter plots for the three locations. Summarize your observations.


````python
plt.figure(figsize=(14, 6))

for i, scn in enumerate(scenarios):
    df_fix = scn.df_fix
    df_status = scn.df_status

    used_in_fix = []
    avg_snr = []
    prev_fix_tstamp = 0
    for fix_tstamp in df_fix["UnixTimeMillis"].unique():
        # Filter status msgs between consecutive fix msgs
        filt = df_status.loc[
            (df_status["UnixTimeMillis"] <= fix_tstamp)
            & (df_status["UnixTimeMillis"] > prev_fix_tstamp)
        ]
        prev_fix_tstamp = fix_tstamp
        if filt.empty:
            continue

        # Find timestamp of status msgs used for current fix
        idx = filt.index[0]
        status_tstamp = df_status.iloc[idx]["UnixTimeMillis"]

        # Filter status msgs by timestamp and the ones used in fix
        df_status_used = df_status[df_status["UnixTimeMillis"] == status_tstamp]
        df_status_used = df_status_used[df_status_used["UsedInFix"] == 1]

        # Fetch error of current fix msg from ground truth
        err = df_fix[df_fix["UnixTimeMillis"] == fix_tstamp]["HDistErrorKm"]
        err = float(err.iloc[0])

        # Save number of satellites used in fix
        used_in_fix.append([err, len(df_status_used)])
        # Compute and save SNR across satellites used in fix
        avg_snr.append([err, df_status_used["Cn0DbHz"].mean()])

    used_in_fix = np.array(used_in_fix)
    plt.subplot(1, 2, 1)
    plt.scatter(
        used_in_fix[:, 0], used_in_fix[:, 1], label=scn.name.title(), color=cmap[i]
    )
    median = np.median(used_in_fix[:, 1])
    plt.axhline(median, linestyle="--", color=cmap[i])

    plt.text(0.0065, 33 - i * 2, f"-- Median = {int(median)}", color=cmap[i])

    avg_snr = np.array(avg_snr)
    plt.subplot(1, 2, 2)
    plt.scatter(avg_snr[:, 0], avg_snr[:, 1], label=scn.name.title())


plt.subplot(1, 2, 1)
plt.title("Satellites Used in Fix v/s Distance Error", fontweight="bold")
plt.xlabel("Error (km)")
plt.ylabel("# of satellites")
plt.legend()

plt.subplot(1, 2, 2)
plt.title("Average SNR v/s Distance Error", fontweight="bold")
plt.xlabel("Error (km)")
plt.ylabel("SNR (dB)")
plt.legend()
plt.show()
````

**Approach**
- For each scenario,    
    - The `Status` messages between successive `Fix` messages are filtered using the timestamp field and the number of satellites used in fix is counted using the `UsedInFix` entry.
    - The average SNR of each satellite signal used for a fix is averaged and saved.
    - The variation of error in estimate is analysed with respect to number of satellites used in fix and the average SNR.

**Observations**
- Satellites used in Fix v/s Distance Error
    - Across the `Open` and `Partial` visibility scenarios, we see a trend where the error in estimate is relatively less as more number of satellites are used for fix.
    - With `Limited` sky exposure, the relative error in fix estimate is the least with the least number of satellites.

- Average SNR v/s Distance Error
    - Across the `Open` and `Partial` visibility scenarios, we see a trend where the error in estimate is relatively less as the average SNR of the signals used in fix increases.
    - With `Limited` sky exposure, the relative error in fix estimate is less with the least SNR value.

**Inference**
- Satellites used in Fix v/s Distance Error
    - Generally, there is a negative correlation between the number of satellites used and the error in position estimate. As the number of satellites used in a GNSS fix increases, the accuracy typically improves in a non-linear fashion. This is because each additional satellite provides more information to help refine the fix, but there are diminishing returns as the number of satellites increases.
    - In some cases, using too many satellites can actually reduce the accuracy of the fix due to increased noise or other errors in the satellite signals. There is typically an optimal number of satellites that provides the best balance between accuracy and reliability.
    - With `Limited` sky exposure, the number of satellites used for fix are significantly less and when the average across measurements is used as the ground truth, this results in low variance of the distribution leading to relatively lower error in estimate.

- Average SNR v/s Distance Error
    - SNR values of satellite signals can be influenced by a multitude of factors like distance from the satellite, Atmospheric quality (weather conditions), LOS obstructions etc.
    - Generally, there is a negative correlation between the Average SNR of the satellite signals and the error in positition estimate. As the signal strength increases, so does the accuracy in the estimate (less error).
    - With `Limited` sky exposure, the average SNR drops significantly but the error is registered low due to signficantly less fix messages leading to low variance in error distribution (since mean is used as ground truth).

````python
# Status msgs from `Partial` scenario
df_status = scenarios[1].df_status
# Filter status msgs used for fix
df_used = df_status[df_status["UsedInFix"] == 1]
# Filter fields of interest
df_sat = df_used.loc[:, ["Svid", "AzimuthDegrees", "Cn0DbHz"]]
# Group by satellite ID and average the rest of the fields for each parameter
df_sat = df_sat.groupby(["Svid"]).mean()

# SNR value along radius
r = df_sat.loc[:, "Cn0DbHz"].to_numpy()
# Azimuth along angle
theta = np.deg2rad(df_sat.loc[:, "AzimuthDegrees"].to_numpy())


plt.figure(figsize=(8, 8))

fig = plt.gcf()

axes_coords = [0.1, 0.1, 0.8, 0.8]

ax_image = fig.add_axes(axes_coords)
ax_image.imshow(mpimg.imread("./images/godav.png"), alpha=0.6, aspect="auto")
ax_image.axis("off")

ax_polar = fig.add_axes(axes_coords, projection="polar")
ax_polar.patch.set_alpha(0)
ax_polar.set_theta_direction(-1)
ax_polar.scatter(theta, r, color="k", marker="^")
ax_polar.grid(True)
ax_polar.set_xlabel("\nRadial: SNR (dB)")
ax_polar.set_ylabel("Angular: Azimuth (degrees)\n\n\n")
plt.suptitle("SNR by Azimuth in Angular Space with Satellite Image", fontweight="bold")
plt.show()
````
